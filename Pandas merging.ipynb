### Merging

```python
d = {'name':['vani','vasu','vamsi','rani'],
    'fruit':['apple','orange','grape','mango'],
    'color':['red','blue','pink','black']}
print(d)
```

```python
import numpy as np
import pandas as pd
df1 = pd.DataFrame(d)
print(df1)
```

```python
d2 = {'name':['a','b','c'],'color':['yellow','red','voilet'],
      'age':[56,67,23]}
df2 = pd.DataFrame(d2)
print(df2)
```

```python
df1.append(df2)
```

```python
print(df1)
```

```python
print(df2)
```

```python
# concatenate
pd.concat([df1,df2],axis = 1)  # axis = 1 --> merge using columns
```

```python
pd.concat([df1,df2],axis=0)  # merging using row wise
```
