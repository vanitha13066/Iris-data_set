- Matplotlib
--------------------------------

  - It is a cross platform data visulization library for making
          2d plots to given data
          
    - Types of plots:
            - Line plot
            - Bar graph
            - Scatter plot
            - Area plot
            - Pie Chart

```python
#line plot
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[2,4,6,8])
plt.title('xy graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Bargraph is used to create bar chart
marks=[88,94,99,92,91]
names=['c','java','python','c++','oracle']
plt.bar(names,marks,color=['red','yellow','black','pink','green'],width=0.6)
```

```python
#Scatter plot
plt.scatter([1,2,3,4],[2,4,6,8],color='b')
plt.show()
```

```python
#Area plot
days=[1,2,3,4]
working=[8,11,9,10]
sleeping=[8,10,11,12]
plt.stackplot(days,working,sleeping,colors=['r','g'],labels=['work','rest'])
plt.legend()
plt.show()
```

```python
#pie chart
rating=[5,3,4.5,4.8]
names=['python','java','c++','c']
plt.pie(rating,labels=names)
plt.legend()
plt.show
```
